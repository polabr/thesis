```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic_2d
import plotly.graph_objects as go
```

```python
lProtonMom = [1082.3474115647148, 1071.660718922391, 1035.984866522877, 1755.4979390211047, 2071.4381320219422, 1724.6150892097285, 1022.7968609775113]
muonMom = [537.6969768331487, 293.46529312479237, 1077.103454886876, 406.41935358331614, 725.2669854539085, 241.65249820876352, 595.9535517314805]
pionMom = [369.5023920459552, 186.1072744772882, 144.87020972639905, 256.1807678197516, 807.807803841935, 494.8852540237025, 320.36656059130286]
```

```python
ax = plt.figure().gca()
ax.yaxis.get_major_locator().set_params(integer=True)

counts, bins = np.histogram(muonMom)
plt.stairs(counts, bins)
plt.xlabel('Muon Momentum (MeV)')
plt.ylabel('Counts')
plt.title('Muon Momentum Distribution')
```
